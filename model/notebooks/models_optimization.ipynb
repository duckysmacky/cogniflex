```python
import torch
import os
import shutil
import gzip
from safetensors.torch import save_file, load_file
```

```python
model_texts = torch.load('best_roberta_fc.pth', map_location=torch.device('cpu'))
model_img_faces = torch.load('resnet_faces88.pth', map_location=torch.device('cpu'))
model_img_general = torch.load('resnet_general92.pth', map_location=torch.device('cpu'))
```

```python
for path in ['best_roberta_fc.pth', 'resnet_faces88.pth', 'resnet_general92.pth']: 
    print(round(os.path.getsize(path)/1024), 'Mb')
```

```python
#turning float32 weights into float16
for model in [model_texts, model_img_faces, model_img_general]:

    for k in model:
        if model[k].dtype == torch.float32:
            model[k] = model[k].half()
```

```python
save_file(model_texts, 'model_texts.safetensors')
save_file(model_img_faces, 'model_img_faces.safetensors')
save_file(model_img_general, 'model_img_general.safetensor')
```

```python
for path in ["model_texts.safetensors", "model_img_faces.safetensors", "model_img_general.safetensors"]: 
    print(round(os.path.getsize(path)/1024), 'Mb')
```
